```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.ticker import PercentFormatter


matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
with open("nobel.csv", 'r') as f:
    Data = pd.read_csv(f)
    Data['usa_born_winner'] = Data['birth_country'] == 'United States of America'
    Data['decade'] = (np.floor(Data['year']/10)*10).astype(int)
    prop_usa_winners = Data.groupby('decade', as_index=True)['usa_born_winner'].mean()
    # print(prop_usa_winners.head(50)) 

    sns.set()
    # and setting the size of all plots.
    plt.rcParams['figure.figsize'] = [11, 7]
    print(plt.rcParams)
    # Plotting USA born winners 
    ax = sns.lineplot(x = 'decade', y = 'usa_born_winner', data = Data)

    # Adding %-formatting to the y-axis
    from matplotlib.ticker import PercentFormatter
    ax.yaxis.set_major_formatter(PercentFormatter(1.0))
    






```
